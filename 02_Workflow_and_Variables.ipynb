# Module 2: The Workflow

**Objective:** Understand how Python "remembers" data and how to fix your notebook when things get out of sync.

### Key Concepts
1. **Variables:** Containers for storing information.
2. **State:** What the computer currently knows (which might be different from what you see).
3. **Restarting:** How to wipe the slate clean.

## Part 1: Variables (The Bucket)
Think of a variable as a bucket with a label. You can put data in it, and the computer will remember it by the label.

**Action:** Run the cell below to create a "bucket" named `revenue`.

```python
revenue = 1000
print("Revenue is stored.")
```

**Guided Discovery:**
Now that the computer knows what `revenue` is, we can use it in other cells.

1. Run the code below.
2. Go back to **Cell 3** (above) and change `1000` to `5000`.
3. Run **Cell 3** again (to update the memory).
4. Run **Cell 5** (below) again. Did the profit update?

```python
costs = 400
profit = revenue - costs
print("The profit is:", profit)
```

## Part 2: The Time Traveler (Debug This)
This is the most common error in notebooks. We read notebooks from **Top to Bottom**, but we can run cells in **Any Order**.

If you run cells out of order, you can confuse the computer (and yourself).

**The Scenario:**
The code below tries to calculate a stock return, but the variable `final_price` is defined *after* the calculation.

**Your Task:**
1. Run the code to see the `NameError`.
2. Use the **Up Arrow** (on the right side of the second cell) to move the `final_price` cell *above* the calculation cell.
3. Run them in the correct order to fix the error.

```python
# CODE BLOCK B (The Data)
initial_price = 100
final_price = 110
```

```python
# CODE BLOCK A (The Calculation)
return_on_investment = (final_price - initial_price) / initial_price
print("The return is:", return_on_investment)
```

## Part 3: The Nuclear Option (Restart Runtime)
Sometimes, your notebook gets so confused (old variables mixing with new ones) that you need a fresh start.

**Action:**
1. Go to the top menu: **Runtime** > **Restart Session** (or Restart Runtime).
2. Click **Yes**.
3. Try to run **Cell 5** immediately.

**What happened?**
It crashed! Why? Because "Restarting" wipes the computer's memory. You must run the cells from the top again to re-define the variables.
