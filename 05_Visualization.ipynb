**Objective:** Turn data into insights using Python's plotting tools.

### Key Concepts
1. **Matplotlib:** The grandfather of all Python plotting libraries.
2. **X and Y:** Understanding how to pass data to a chart.
3. **Exporting:** Saving your plot as an image file.

## Part 1: The Basic Plot
We usually import `matplotlib.pyplot` and nickname it `plt`.

**Action:**
Run the code to generate a simple line chart representing a stock's growth.

```python
import matplotlib.pyplot as plt

# Data
days = [1, 2, 3, 4, 5]
prices = [100, 102, 101, 105, 110]

# The Plot Command
plt.plot(days, prices)
plt.title("Simulated Stock Price")
plt.xlabel("Day")
plt.ylabel("Price ($)")
plt.show()
```

## Part 2: Guided Discovery (Styles)
The default line is blue and simple. Let's make it look professional.

**Action:**
1. Run the code below to see a dashed red line with markers.
2. **Experiment:** Change `'r--o'` to `'g-^'` (green, solid line, triangle markers).

```python
plt.plot(days, prices, 'r--o') # r=red, --=dashed, o=dots
plt.grid(True) # Add a grid for readability
plt.show()
```

## Part 3: Debug This (Dimension Mismatch)
A plot needs an X for every Y. If the lists are different lengths, Python gets confused.

**The Scenario:**
We have 5 days but only 4 price points.

**Your Task:**
1. Run the code to see the `ValueError`.
2. Read the error: `x and y must have same first dimension`.
3. **Fix it:** Add a 5th price (e.g., `108`) to the `prices` list so it matches the `days` list.

```python
days = [1, 2, 3, 4, 5]
prices = [100, 102, 101, 105] # Oops, missing one!

plt.plot(days, prices)
plt.show()
```

## Part 4: Exporting for the Book
You can save your plot as a file to download or put in a report.

**Action:**
1. Run the code below.
2. Look at the **Files** sidebar on the left (Folder icon).
3. You will see `my_chart.png`. You can double-click to view it or right-click to download.

```python
plt.plot(days, [100, 105, 110, 115, 120])
plt.title("Growth Chart")

# Save the figure
plt.savefig("my_chart.png", dpi=300) # dpi=300 makes it high quality for printing
print("Chart saved to Files sidebar!")
```
